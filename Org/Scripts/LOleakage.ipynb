```python
import numpy as np
import math
import matplotlib.pyplot as plt

#variables
Li=10**(-10/10)
Lq=10**(-13/10)
Pi=0
Pq=0

#parametric plots
numSamples=1000
Idc = np.linspace(start=-0.5, stop=0.5, num=numSamples) ## Attenuation dB
Qdc= np.linspace(start=-0.5, stop=0.5, num=numSamples) ## 

I, Q = np.meshgrid(Idc,Qdc)

#Calculation
z1=I*Q+I*Lq*np.cos(Pq)+Q*Li*np.cos(Pi)+Li*Lq*np.cos(Pi-Pq)  #phase condition hyperbola
z2=I*I-Q*Q+2*I*Li*np.cos(Pi)-2*Q*Lq*np.cos(Pq)+Li*Li-Lq*Lq  #amplitude condition hyperbola

c1 = plt.contour(Idc,Qdc,z1,[0],colors='red')
c2 = plt.contour(Idc,Qdc,z2,[0],colors='blue')
c = plt.contourf(Idc,Qdc,10*np.log10(np.abs(z1)+np.abs(z2)),[-50,-45,-40,-35,-30,-25,-20,-15,-10,-5,0])
plt.colorbar()
cc = plt.contour(Idc,Qdc,np.abs(z1)+np.abs(z2),[0.0005],colors='black')
h1,_ = c1.legend_elements()
h2,_ = c2.legend_elements()
plt.legend([h1[0], h2[0]], ['Amplitude','Phase'])
plt.xlabel("$I_{DC}$")
plt.ylabel("$Q_{DC}$")
plt.title("LO Leakage Level - Cero Phase")
plt.savefig('ceroLOleakagehyperbolesceropha.pdf')
plt.show()
```
