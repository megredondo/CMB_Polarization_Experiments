```python
# import all necessary modules here
import numpy as np
import math
import matplotlib.pyplot as plt

#parametric plots
numSamples=30
ampim = np.linspace(start=1, stop=1.5, num=numSamples) ## Amplitude Imbalance
phaseim = np.linspace(start=0.1, stop=20, num=numSamples) ## Phase Imbalance
ampimdb = 10*np.log10(ampim) ## Amplitude Imbalance dB
x, y = np.meshgrid(ampim,phaseim)

#Power calculations
IMRR=((x**2)+1-2*x*np.cos(y*np.pi/180))/((x**2)+1+2*x*np.cos(y*np.pi/180))    #image rejection ratio
IMRRdb=10*np.log10(IMRR)                   #image rejection ratio in db

h = plt.contourf(phaseim,ampimdb,IMRRdb,[-60,-55,-50,-45,-40,-35,-30,-25,-20,-15,-10]) 
plt.colorbar()
plt.ylabel("Amplitude Imbalance [dB]")
plt.xlabel("Phase Imbalance [Â°]")
plt.title("IMRR [dB]")
plt.savefig('IMRR.pdf')
plt.show()
```
