```python
import numpy as np
import math
import matplotlib.pyplot as plt
from matplotlib import ticker, cm

#parametric plots
Ta=290
Tc=10
numSamples=30
Atdb = np.linspace(start=0, stop=60, num=numSamples) ## Attenuation dB
dtdb= np.linspace(start=-60, stop=0, num=numSamples) ## 

At=10**(Atdb/10) ## Hot Attenuation times
diftemp=10**(dtdb/10) # Cold Attenuation times

x, y = np.meshgrid(diftemp,At)

#Power calculations
gain=((1/x)+(y-1))/y
gaindb=10*np.log10(gain)

c = plt.imshow(gaindb,extent=[-60, 0, 0, 60],origin='lower') 
plt.colorbar()
plt.ylabel("Cold Attenuation [dB]")
plt.xlabel("Temperature Ratio [Times]")
plt.title("Noise Reduction Factor [dB]")
plt.savefig('NoiseReductionFactor.pdf')
plt.show()

c = plt.contourf(dtdb,Atdb,gaindb,origin='lower') 
plt.colorbar()
plt.ylabel("Cold Attenuation [dB]")
plt.xlabel("Temperature Ratio [Times]")
plt.title("Noise Reduction Factor [dB]")
plt.savefig('NoiseReductionFactor.pdf')
plt.show()
```
