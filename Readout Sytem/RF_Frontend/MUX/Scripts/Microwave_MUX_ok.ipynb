```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Universal Parameters and constants
h=6.62607004e-34;           #Plank Constant
e=9.1093837015e-31;         #Electron Mass
Phi0=h/(2*e);               #Quantum Magnetic Flux
```

```python
# SQUID Parameters 
lamb=1/3;                 #Ls/Lj (Josepson inductance Lj=h_bar/2eIc)
Ls=18.9e-12;                #Squid Self inductance
Lj=lamb*Ls;               #(Josepson inductance Lj=h_bar/2eIc)       
Lc=77.6e-12;                #Screening Inductance
Mc=1.65e-12;                #Screening Mutual Inductance
```

```python
# Parameters Resonator
f1=6e9;               #Unperturbed Resonance Frequency
Z1=50;                #Resonator Impedance
Cc=1e-15;             #Coupling Capacitance
```

```python
N=200;  #Number of points
M=1;   #Number of Quantum Magnetic Fluxes  
flux=np.arange(-M,M,M/N);        #Flux Vector
fr=f1-4*(f1**2)*Cc*Z1-4*(f1**2)*Lc/Z1+4*(f1**2)*lamb*Mc**2*np.cos(2*np.pi*flux)/(Z1*Ls);   #Resonance frequency with flux
fsh=4*(f1**2)*lamb*Mc**2*np.cos(2*np.pi*flux)/(Z1*Ls);   #Resonance frequency shift with flux
```

```python
#Ploting Frequency Vs. Flux

fig, ax = plt.subplots()
ax.plot(flux,fr)

ax.set(xlabel='Flux [Phi0]', ylabel='Frequency[GHz]',
       title='Resonance Frequency Vs. Flux')
ax.grid()

fig.savefig("Frvsflux.pdf")
plt.show()
```

```python
#Ploting Frequency Vs. Flux

fig, ax = plt.subplots()
ax.plot(flux,fsh/1e3)

ax.set(xlabel='Flux [Phi0]', ylabel='Frequency Shift [KHz]',
       title='Resonance Frequency Shift Vs. Flux')
ax.grid()

fig.savefig("Fshvsflux.pdf")
plt.show()
```
