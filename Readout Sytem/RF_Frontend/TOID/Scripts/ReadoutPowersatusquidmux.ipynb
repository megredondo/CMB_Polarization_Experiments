```python
# import all necessary modules here
import numpy as np
import math
import matplotlib.pyplot as plt
#variables
p=1				        #flux amplitude
phi0=2e-15				#magnetic quantum flux	
Mc=5e-12				  #mutual inductance beweeen resonator and squid
Z1=50             #characteristic impedance
Qc=10000          #carged quality factor

#parametric plots
numSamples=1000
px = np.linspace(start=0.1*p, stop=10*p, num=numSamples).reshape((numSamples,1)) ## Inputs
Mcx = np.linspace(start=0.1*Mc, stop=10*Mc, num=numSamples).reshape((numSamples,1)) ## Inputs
Z1x = np.linspace(start=Z1, stop=2*Z1, num=numSamples).reshape((numSamples,1)) ## Inputs
Qcx = np.linspace(start=0.1*Qc, stop=10*Qc, num=numSamples).reshape((numSamples,1)) ## Inputs

Pina=(np.pi*Z1*px*px*phi0*phi0)/(64*Qc*Mc**2)    #power in watts
Pinadb=10*np.log10(Pina/1e-3)
Pinb=(np.pi*Z1*p*p*phi0*phi0)/(64*Qc*Mcx**2)    #power in watts
Pinbdb=10*np.log10(Pinb/1e-3)
Pinc=(np.pi*Z1x*p*p*phi0*phi0)/(64*Qc*Mc**2)    #power in watts
Pincdb=10*np.log10(Pinc/1e-3)
Pind=(np.pi*Z1*p*p*phi0*phi0)/(64*Qcx*Mc**2)    #power in watts
Pinddb=10*np.log10(Pind/1e-3)


#Power calculations
Pin=(np.pi*Z1*p*p*phi0*phi0)/(64*Qc*Mc**2)    #power in watts
Pindbm=10*np.log10(Pin/1e-3)                    #power in dbm
print("Pin",round(Pin,12))						          #show results
print("Pindbm",round(Pindbm,3))

plt.figure(1)
plt.plot(px,Pinadb)
plt.ylabel("Pin[dbm]")
plt.xlabel("p")
plt.title("Pin vs. p")
plt.savefig('Pinvsp.png')

plt.figure(2)
plt.plot(px,Pinbdb)
plt.ylabel("Pin[dbm]")
plt.xlabel("Mc")
plt.title("Pin vs. Mc")
plt.savefig('PinvsMc.png')

plt.figure(3)
plt.plot(px,Pincdb)
plt.ylabel("Pin[dbm]")
plt.xlabel("Z1")
plt.title("Pin vs. Z1")
plt.savefig('PinvsZ1.png')

plt.figure(4)
plt.plot(px,Pinddb)
plt.ylabel("Pin[dbm]")
plt.xlabel("Qc")
plt.title("Pin vs. Qc")
plt.savefig('PinvsQc.png')

```
